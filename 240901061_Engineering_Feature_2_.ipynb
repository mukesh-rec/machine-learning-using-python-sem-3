```python
import pandas as pd
import numpy as np

# Creating a dummy DataFrame of 12 numbers randomly ranging from 150-180 for height
df = pd.DataFrame({'Height': [150.4, 157.6, 170, 176, 164.2, 155,
                             159.2, 175, 162.4, 176, 153, 170]})

# Printing DataFrame before sorting continuous to categories
print("Before:")
print(df)

# Categorizing Height into 3 Categories
# Short: (150,157]
# Average: (157,169]
# Tall: (169,180]

df['Label'] = pd.cut(df['Height'],
                     bins=[150, 157, 169, 180],
                     labels=['Short', 'Average', 'Tall'])

# Printing DataFrame after categorization
print("After:")
print(df)

# Check number of values in each bin
print("Categories:")
print(df['Label'].value_counts())
```
