```python
import pandas as pd
import numpy as np

# Creating a dummy DataFrame of 15 numbers randomly ranging from 1-100 for age
df = pd.DataFrame({'Age': [42, 15, 67, 55, 1, 29, 75, 89, 4,
                           10, 15, 38, 22, 77]})

# Printing DataFrame before sorting continuous to categories
print("Before:\n")
print(df)

# Categorizing Age into 4 categories
# Baby/Toddler: (0-3]
# Child: (3,17]
# Adult: (17,63]
# Elderly: (63,99]

df['Label'] = pd.cut(df['Age'],
                     bins=[0, 3, 17, 63, 99],
                     labels=['Baby/Toddler', 'Child', 'Adult', 'Elderly'])

# Printing DataFrame after categorization
print("After:\n")
print(df)

# Check the number of values in each bin
print("Categories:\n")
print(df['Label'].value_counts())
```
