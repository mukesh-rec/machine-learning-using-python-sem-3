```python
import matplotlib.pyplot as plt

# Data
days = [1, 2, 3, 4, 5]

bmw = [50, 40, 70, 80, 20]
audi = [80, 20, 20, 50, 60]

# Bar positions
x1 = [d - 0.2 for d in days]
x2 = [d + 0.2 for d in days]

# Plot bars
plt.bar(x1, bmw, width=0.4, label="BMW", color='blue')
plt.bar(x2, audi, width=0.4, label="Audi", color='cyan')

# Labels and title
plt.xlabel('Days')
plt.ylabel('Distance (kms)')
plt.title('Information')

# X-axis ticks
plt.xticks(days)

# Legend
plt.legend()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(300, 20, 250)
plt.hist(x, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([40, 30, 20, 10])
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'lightgreen', 'lightblue', 'pink']

plt.pie(y, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ControlSystem", "PSA"]

plt.pie(y, labels=mylabels, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
plt.title("Distribution of Subjects")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ControlSystem", "PSA"]
myexplode = [0, 0, 0.3, 0]  # Third slice is exploded

plt.pie(y, labels=mylabels, explode=myexplode, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
plt.title("Subjects Distribution with Explode")
plt.show()
```
