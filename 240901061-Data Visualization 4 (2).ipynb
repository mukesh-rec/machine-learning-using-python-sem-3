```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

#3-D plot
X=np.array([1,2,6,8,12,14,20])
Y=np.array([3,8,1,10,12,16,18])
Z=np.array([5,10,15,20,25,30,35]) # Added Z array for 3D plot
Fig=plt.figure(figsize=(12,8))
Ax=plt.axes(projection='3d') # Corrected 'prohection' to 'projection'
Plt_3d=Ax.scatter3D(X,Y,Z) # Changed 'ax' to 'Ax', 'x' to 'X', 'y' to 'Y', added Z
Fig.colorbar(Plt_3d) # Changed 'Plt' to 'Fig' for colorbar
plt.show()

#plotly package
X_plotly=np.array([10,80,100,120,150,180,250]) # Renamed X to X_plotly to avoid confusion

fig=px.scatter(x=X_plotly) # Changed 'x' to 'X_plotly'
fig.show()
```
